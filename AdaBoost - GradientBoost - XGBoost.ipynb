```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
dataset=dataset.astype(int)
dataset.head()
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
independent.head()
```

```python
dependent.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.2, random_state=0)
```

```python
x_train.head()
```

```python
x_train.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.ensemble import AdaBoostRegressor
Regressor_ABR = AdaBoostRegressor()
Regressor_ABR = Regressor_ABR.fit(x_train, y_train)
```

```python
y_pred_ABR = Regressor_ABR.predict(x_test)
```

```python
from sklearn.metrics import r2_score
R2_score_ABR = r2_score(y_test, y_pred_ABR)
print(R2_score_ABR)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
Regressor_GBR = GradientBoostingRegressor()
Regressor_GBR = Regressor_GBR.fit(x_train, y_train)
```

```python
y_pred_GBR = Regressor_GBR.predict(x_test)
```

```python
R2_score_GBR = r2_score(y_test, y_pred_GBR)
print(R2_score_GBR)
```
